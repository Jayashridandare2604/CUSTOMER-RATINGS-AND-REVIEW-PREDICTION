#### Scraping laptop's ratings and reviews

```python
# lets first connect to the web driver
PATH = r"C:\Users\JAYASHRI\Downloads\chromedriver_win32 (1)\chromedriver.exe" #extracting path of the webdriver
driver = webdriver.Chrome(PATH)
driver.maximize_window()
time.sleep(2)

# opening flipcart review page 
url = "https://www.flipkart.com/apple-macbook-air-m1-8-gb-256-gb-ssd-mac-os-big-sur-mgn63hn-a/product-reviews/itmde54f026889ce?pid=COMFXEKMGNHZYFH9&lid=LSTCOMFXEKMGNHZYFH9EOWT4E&marketplace=FLIPKART&page=1"
driver.get(url)
time.sleep(2)
```

```python
Review1=[]
Rating1= []

review1=driver.find_elements_by_xpath('//div[@class="t-ZTKy"]/div/div')
for i in review1:
        
    Review1.append(i.text)
              
rating1=driver.find_elements_by_xpath('//div[@class="_3LWZlK _1BLPMq"]')
for j in rating1:
        
    Rating1.append(j.text)
```

```python
# Clicking on next button
nxt_button=driver.find_element_by_xpath('//a[@class="_1LKTO3"]')
time.sleep(2)
nxt_button.click()
```

```python
start=0
end=70

for page in range(start,end):
    
    # Extracting Review from xpath
    review1=driver.find_elements_by_xpath('//div[@class="t-ZTKy"]/div/div')
    for j in review1:
        Review1.append(j.text)
        
    # Extracting Rating from xpath    
    rating1=driver.find_elements_by_xpath('//div[(@class="_3LWZlK _1BLPMq" or @class="_3LWZlK _32lA32 _1BLPMq")or @class="_3LWZlK _1rdVr6 _1BLPMq" ]')
    for j in rating1:                             
        Rating1.append(j.text)
    time.sleep(1)
    
    nxt_button=driver.find_element_by_xpath('//a[@class="_1LKTO3"][2]')     
    nxt_button.click()
    time.sleep(1)
```

```python
driver.close()
```

```python
# Creating the dataframe from the scraped data
Apple_Laptop=pd.DataFrame({"Review":Review1,"Rating":Rating1})

Apple_Laptop
```
